## Scraping raw web data

```python
# import some libraries
import requests
from bs4 import BeautifulSoup
```

```python
# what URL do we want to scrape?
url = "https://www.yelp.com/biz/m-sushi-durham"

# get that page information (i.e., what you would
# use to render it in ypur browser)
page = requests.get(url)

# parse the page content as html
soup = BeautifulSoup(page.content, "html.parser")

# What does this object look like?
soup
```

```python
url = "https://www.yelp.com/biz/m-sushi-durham?start=%d"

starting_point = 0
keep_going = True

df = pd.DataFrame(columns=["text", "rating", "date"])
while keep_going:
    page = requests.get(url % starting_point)
    starting_point += 20
    keep_going = False
    
    soup = BeautifulSoup(page.content, "html.parser")

    reviews = soup.findAll("li", {"class":"lemon--li__373c0__1r9wz"})
    
    for review in reviews:
        text = review.findAll("span", {"lang":"en"})
        review_and_metadata = str(review)
    
        date = review.findAll("span", {"class":"text-color--mid__373c0__3G312"})
    
        stars = 0
        if "5 star rating" in review_and_metadata:
            stars = 5
        if "4 star rating" in review_and_metadata:
            stars = 4
        if "3 star rating" in review_and_metadata:
            stars = 3
        if "2 star rating" in review_and_metadata:
            stars = 2
        if "1 star rating" in review_and_metadata:
            stars = 1

        if len(text) != 0:
            df = df.append({"text": text[0], "rating": stars, "date": date[0].contents[0]}, ignore_index=True)
            keep_going = True
    
df
```

## But sometimes there's an easier way...

```python
url = "https://www.yelp.com/biz/m-sushi-durham"

page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

data = soup.findAll("script", {"type":"application/ld+json"})
for d in data:
    if 'aggregateRating' in d.contents[0]:
        print(d.contents[0])
```

```python
import json
parsed_json = (json.loads(d.contents[0]))
reviews = parsed_json['review']
reviews
```

```python
df = pd.DataFrame(columns=["text", "rating", "date", "author"])
for review in reviews:
    df = df.append({"text": review["description"],
                    "rating": review['reviewRating']['ratingValue'],
                    "date": review['datePublished'],
                    "author": review['author']}, ignore_index=True)
df
```

```python
import time
import numpy as np
url = "https://www.yelp.com/biz/m-sushi-durham?start=%d"

starting_point = 0
keep_going = True

df = pd.DataFrame(columns=["text", "rating", "date", "author"])
while keep_going:
    page = requests.get(url % starting_point)
    starting_point += 20
    keep_going = False
    time.sleep(np.random.randint(20))
    
    soup = BeautifulSoup(page.content, "html.parser")
    data = soup.findAll("script", {"type":"application/ld+json"})
    for d in data:
        if 'aggregateRating' in d.contents[0]:
            parsed_json = (json.loads(d.contents[0]))
            reviews = parsed_json['review']
            reviews
            
            for review in reviews:
                df = df.append({"text": review["description"],
                    "rating": review['reviewRating']['ratingValue'],
                    "date": review['datePublished'],
                    "author": review['author']}, ignore_index=True)
                keep_going = True
df
```

### "crawling" from page to page

```python
url = "https://www.yelp.com/biz/m-sushi-durham"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
links = soup.findAll("a")
print(len(links))
print(links)
```

```python
for link in links:
    if 'href' in str(link) and 'related_bizes' in link['href']:
        print(link['href'])
```

```python
import numpy as np

def scrape_page(url):
    print("scraping page", url)
    return(["url %d" % np.random.randint(16),"url %d" % np.random.randint(16)])

pages_to_scrape = ['my starting url']
pages_scraped = set()
while pages_to_scrape:
    page = pages_to_scrape.pop(0)
    
    new_pages = scrape_page(page)
    pages_scraped.add(page)
    print("  found pages", new_pages)
    
    for new_page in new_pages:
        if new_page not in pages_scraped and new_page not in pages_to_scrape:
            pages_to_scrape.append(new_page)
            print("   adding page", new_page)
            
    print("%d pages scraped, %d to go!" % (len(pages_scraped), len(pages_to_scrape)))
    
print('all done!')
```
